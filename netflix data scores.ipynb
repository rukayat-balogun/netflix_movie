```python
read_file["runtime/imdb_score"] = round(read_file['Runtime'] / read_file["IMDB Score"],0)
```

```python
read_file
```

```python
color_code = (200/255,82/255,0/255)
fig = plt.figure(figsize=(15,9))
plt.plot(read_file["Premiere"], read_file["runtime/imdb_score"], c= color_code)
plt.xlabel("Premiere")
plt.ylabel("runtime/imdb_score")
plt.title("Premier of movie Vs runtime/imdb_score")
plt.xticks(rotation=45)
plt.show()
```
